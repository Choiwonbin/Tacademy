## Matplotlib
- Reference: https://matplotlib.org/gallery/index.html
- Tutorial: http://www.labri.fr/perso/nrougier/teaching/matplotlib/

```python
# 필요한 library import
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 그래프 바로 확인
%matplotlib inline
```

```python
# 그래프 그릴 x, y 지정
x = np.linspace(0,5,11)
y = x ** 2
x, y
```

### Functional Method

```python
# 그래프 그리기
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
# Create Multi-plots
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

### Object-oriented Method

```python
# create figure object (empty canvas)
fig = plt.figure()

# 축 지정 ([left, bottom, width, height])
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8 ])

# 그래프 그리기
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
# 큰 그래프 안에 작은 그래프 넣기
# create figure object (empty canvas)
fig = plt.figure()

# 축 지정 ([left, bottom, width, height])
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

# 그래프 그리기
ax1.plot(x,y)
ax2.plot(y,x)
ax1.set_title('Larger Plot')
ax2.set_title('Smaller Title')
```

```python
# 그래프의 사이즈(figsize), 해상도(dpi) 조절
fig = plt.figure(figsize = (8,2), dpi = 60)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
# 그래프 저장
fig.savefig('my_picture.png', dpi = 200)
```

```python
# 범례 지정
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x, x ** 2, label = 'X Squared')
ax.plot(x, x ** 3, label = 'X Cubed')

ax.legend(loc = (0.1, 0.1))
```

```python
# 그래프 모양 지정
fig = plt.figure()
ax = fig.add_axes( [0, 0, 1, 1] )

# color: 그래프 색상, linewidth: 그래프 굵기, alpha: 그래프 투명도, linestyle: 그래프 선
# marker: 마커 모양(o, +, *, 1, 2, etc), markersize: 마커 크기, markerfacecolor: 마커 색상
# markeredgewidth: 마커 선 굵기, markeredgecolr: 마커 선 색상
ax.plot(x, y, color = '#FF8C00', linewidth = 0.5, alpha = 0.5, linestyle = '--',
        marker = 'o', markersize = 10, markerfacecolor = 'yellow',
        markeredgewidth = 3, markeredgecolor = 'green')
```

```python
# 축 범위 지정
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

# lw: linewidth, ls: linestyle
ax.plot(x, y, color = 'purple', lw = 2, ls = '--')

ax.set_xlim([0, 1])
ax.set_ylim([0, 2])
```

## Seaborn
- Reference: https://seaborn.pydata.org/examples/index.html#example-gallery

```python
# 필요한 library import
import seaborn as sns
```

### Distribution Plots

```python
# plot할 data를 tips로 불러오기
tips = sns.load_dataset('tips')
tips.head()
```

```python
# tips의 total_bill column을 기준으로 40개로 구분한 histogram plot
sns.distplot(tips['total_bill'], bins=40)
```

```python
# tips의 total_bill, tip column을 기준으로 scatter, histogram plot
sns.jointplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
# tips의 sex column을 기준으로 scatter, histogram plot
sns.pairplot(tips, hue = 'sex', palette = 'coolwarm')
```

### Categorial Plots

```python
# tips의 sex column을 기준으로 total_bill column의 평균(막대 높이), 편차(에러바) plot
sns.barplot(x = 'sex', y = 'total_bill', data = tips)
```

```python
# tips의 sex column을 기준으로 data의 양 plot
sns.countplot(x = 'sex', data = tips)
```

```python
# tips의 day column을 기준으로 total_bill의 Box-Whisker Plot
sns.boxplot(x = 'day', y = 'total_bill', data = tips)
```

### Matrix Plots

```python
# plot할 데이터를 flights로 불러오기
flights = sns.load_dataset('flights')
flights.head()
```

```python
# tips의 상관관계를 heatmap plot
tc = tips.corr()
sns.heatmap(tc, annot=True, cmap = 'coolwarm')
```

```python
# flights의 month, year column을 기준으로 passengers column의 피벗테이블 생성
fp = flights.pivot_table(index = 'month', columns = 'year', values = 'passengers')
fp
```

```python
# fp의 heatmap plot
sns.heatmap(fp, cmap = 'coolwarm', linecolor = 'white', linewidths = 1)
```

```python
# fp의 clustermap plot
sns.clustermap(fp, cmap = 'coolwarm', standard_scale = 1)
```

### Regression Plots

```python
# tips의 sex column을 기준으로 total_bill, tip column의 regression plot
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex', markers = ['o', 'v'])
```

### Style and Color

```python
# set_style: 배경색(white, dark, etc), set_context: 그래프 서식(poster, talk, etc), despine: 배경 경계선 제거
sns.set_style('white')
sns.set_context('poster')
sns.countplot(x = 'sex', data = tips)
sns.despine()
```

```python
# palette: 색상 (https://matplotlib.org/examples/color/colormaps_reference.html)
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex', palette = 'seismic')
```
